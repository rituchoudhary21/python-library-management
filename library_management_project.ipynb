```python
# Library Management System Welcome Script

def welcome_message():
    print("==========================================")
    print("  Welcome to the Library Management System")
    print("==========================================")
    print("Manage your books, members, and transactions with ease!")


welcome_message()
# Step 1.2: Initial Book Collection
library = {
    "Python": {"author": "Test", "available": True, "issued_to": None},
    "Data Science": {"author": "Test1", "available": False, "issued_to": "none"}
}

# Step 1.2: Predefined User Credentials
credentials = {"admin": "admin123", "librarian": "lib123"}

# Step 1&5: Login Function
def login():
    for _ in range(3):
        username = input("Enter username: ")
        password = input("Enter password: ")
        if username in credentials and credentials[username] == password:
            print(f"Welcome to the Library Management System, {username}!\n")
            return True
        else:
            print("Username or password is invalid.\n")
    print("Too many failed attempts. Exiting.")
    return False

# Step : View Books
def view_books():
    print("\nAvailable Books:")
    for book, details in library.items():
        status = "Available" if details["available"] else f"Issued to {details['issued_to']}"
        print(f"{book} by {details['author']} - {status}")

# Step 2: Add a Book
def add_book():
    title = input("Enter book title: ")
    author = input("Enter book author: ")
    if title in library:
        print("Book already exists.")
    else:
        library[title] = {"author": author, "available": True, "issued_to": None}
        print(f"Book '{title}' added successfully.")

# Step 3: Issue a Book
def issue_book():
    title = input("Enter book title to issue: ")
    if title in library:
        if library[title]["available"]:
            person = input("Enter name of person issuing the book: ")
            library[title]["available"] = False
            library[title]["issued_to"] = person
            print(f"Book '{title}' issued to {person}.")
        else:
            print(f"Book is already issued to {library[title]['issued_to']}.")
    else:
        print("Book not found.")

# Step 4: Return a Book
def return_book():
    title = input("Enter book title to return: ")
    if title in library:
        if not library[title]["available"]:
            library[title]["available"] = True
            library[title]["issued_to"] = None
            print(f"Book '{title}' returned successfully.")
        else:
            print("This book was not issued.")
    else:
        print("Book not found.")

# Step 6: View Issued Books
def view_issued_books():
    print("\nIssued Books:")
    found = False
    for book, details in library.items():
        if not details["available"]:
            print(f"{book} - Issued to {details['issued_to']}")
            found = True
    if not found:
        print("No books are currently issued.")

# Step 7: Main Menu
def menu():
    while True:
        print("\n--- Library Menu ---")
        print("1. View Books")
        print("2. Add a Book")
        print("3. Issue a Book")
        print("4. Return a Book")
        print("5. View Issued Books")
        print("6. Exit")

        try:
            choice = int(input("Enter your choice (1–6): "))
            if choice == 1:
                view_books()
            elif choice == 2:
                add_book()
            elif choice == 3:
                issue_book()
            elif choice == 4:
                return_book()
            elif choice == 5:
                view_issued_books()
            elif choice == 6:
                print("Thank you for using the Library Management System.")
                break
            else:
                print("Invalid choice. Please enter a number between 1 and 6.")
        except ValueError:
            print("Please enter a valid number.")

if login():
    menu()
    
```
